```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
data = pd.read_csv("/Users/sokrates/Desktop/Asil_Lab Assignment/creditcard.csv")
```

```python
data.info()
data.describe()
print(data.columns)

print(data.columns.tolist())
```

```python
sns.countplot(x='Class', data=data)
plt.title('Sınıf Dağılımı (0: Normal, 1: Dolandırıcılık)')
plt.show()

# Oransal olarak bakmak daha çarpıcıdır
print(data['Class'].value_counts(normalize=True) * 100)
```

```python
# Dolandırıcılık ve Normal işlemlerin harcama miktarı farkı
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(12,6))

ax1.hist(data.Amount[data.Class == 1], bins=30)
ax1.set_title('Dolandırıcılık İşlemleri')

ax2.hist(data.Amount[data.Class == 0], bins=30)
ax2.set_title('Normal İşlemler')

plt.xlabel('Miktar ($)')
plt.yscale('log') # Logaritmik ölçek farkı görmeyi kolaylaştırır
plt.show()
```

```python
sns.distplot(data['V17'][data['Class'] == 1], label='Dolandırıcı')
sns.distplot(data['V17'][data['Class'] == 0], label='Normal')
plt.legend()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.gridspec as gridspec

# V sütunlarını seçelim
v_features = data.iloc[:, 1:29].columns 

plt.figure(figsize=(16, 28))
gs = gridspec.GridSpec(7, 4) # 28 değişken için 7x4 matris

for i, col in enumerate(v_features):
    ax = plt.subplot(gs[i])
    # Class 1 (Dolandırıcılık) ve Class 0 (Normal) dağılımları
    sns.kdeplot(data[data.Class == 1][col], shade=True, color="red", label="Fraud", ax=ax)
    sns.kdeplot(data[data.Class == 0][col], shade=True, color="blue", label="Normal", ax=ax)
    ax.set_title(f'Dağılım: {col}')
    ax.set_xlabel('')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.kdeplot(data[data['Class'] == 0]['Amount'], label='Normal', shade=True)
sns.kdeplot(data[data['Class'] == 1]['Amount'], label='Fraud', shade=True)
plt.title('İşlem Tutarı Dağılımı (Sınıflara Göre)')
plt.xlim(0, 2000) 
plt.legend()
plt.show()
```
